```python
import h5py

# 파일 열기
f = h5py.File(r'C:\Users\smart\Desktop\2023-MATLAB-AI-contest\checkpoint.h5', 'r')

# 파일 내의 모든 항목을 재귀적으로 탐색하고 출력하는 함수
def print_attrs(name, obj):
    print(name)
    for key, val in obj.attrs.items():
        print("    %s: %s" % (key, val))

# 위 함수를 사용하여 모든 항목 탐색
f.visititems(print_attrs)
```

```python
import h5py

# h5 파일 열기
with h5py.File(r'C:\Users\smart\Desktop\2023-MATLAB-AI-contest\checkpoint.h5', 'r') as f:
    # 파일 내의 모든 그룹/데이터세트를 재귀적으로 탐색
    def h5py_dataset_iterator(g, prefix=''):
        for key in g.keys():
            item = g[key]
            path = f'{prefix}/{key}'
            if isinstance(item, h5py.Dataset): # item is dataset
                print(path)
                print("Shape: ", item.shape)
                print("Data type: ", item.dtype)
                print("Value: ", item[()])
                print("Attributes: ", dict(item.attrs))
            elif isinstance(item, h5py.Group): # item is group (folder)
                h5py_dataset_iterator(item, path)

    # 위 함수를 사용하여 모든 데이터세트 탐색
    h5py_dataset_iterator(f, '')
```
